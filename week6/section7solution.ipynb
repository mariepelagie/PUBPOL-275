**Exercise 5: Do a moving window regression where you regress average August rainfall on average March rainfall across locations within a window. Use a window that is 21 pixels by 21 pixels (10 pixels on each side of the center pixel) and record the regression coefficient at the location of the center pixel. Plot the results.**

````python
# TODO: 30-min exercise
def rolling_regression(i, j, x, y, window=10):
    assert x.shape == y.shape
    i_min = max(0, i - window)
    j_min = max(0, j - window)
    i_max = min(x.shape[0], i + window + 1)
    j_max = min(x.shape[1], j + window + 1)
    subset_x = x[i_min:i_max, j_min:j_max]
    subset_y = y[i_min:i_max, j_min:j_max]
    if (subset_x.sum() == 0 or subset_y.sum() == 0 or
            np.isnan(subset_x.sum()) or np.isnan(subset_y.sum())):
        return np.nan
    b = linregress(subset_x.flatten(),
                   subset_y.flatten())
    return b[1]


# subset ds
rain3 = ds['rainfall'].sel(month=3).mean(dim='year').values
rain8 = ds['rainfall'].sel(month=8).mean(dim='year').values

# define an empty grid
regression_grid = np.full_like(rain3, np.nan)

# fill the grid
for i in range(regression_grid.shape[0]):
    for j in range(regression_grid.shape[1]):
        regression_grid[i, j] = rolling_regression(i, j, rain3, rain8)

# now take grid and add back in lat/lon coordinates
regression_grid = xr.DataArray(
    regression_grid,
    dims=('lat', 'lon'),
    coords={'lat': ds['lat'], 
            'lon': ds['lon']})

# now plot the beta coefficients as a map
regression_grid.plot()
````
